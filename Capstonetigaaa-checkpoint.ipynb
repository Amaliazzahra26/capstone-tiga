# 3. Data Preprocessing

Tujuan dari data processing adalah untuk menentukan feature enggineringyang tepat terhadap feature - feature yang ada,
sehingga nantinya akan mendapatkan data yang bersih dan siap diolah kedalam machine learning yang akan digunakan nantinya.

```python
listItem = []
for col in df.columns :
    listItem.append([col, df[col].dtype, df[col].isna().sum(), round((df[col].isna().sum()/len(df[col])) * 100,2),
                    df[col].nunique(), list(df[col].drop_duplicates().sample(2).values)]);

dfDesc = pd.DataFrame(columns=['dataFeatures', 'dataType', 'null', 'nullPct', 'unique', 'uniqueSample'],
                     data=listItem)
dfDesc
```
